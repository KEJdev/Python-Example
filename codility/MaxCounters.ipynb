```python
A = [1]
N = 1
```

```python
# O(N + M)
def solution(N, A):
    temp = [0]*N
    df, df2 = 0,0

    for val in A:
        if val == N+1:
            df = df2
        else:
            if temp[val-1] < df:
                temp[val-1] = df+1
            else:
                temp[val-1]+=1
            if temp[val-1] > df2:
                df2 = temp[val-1]
                
    for idx in range(N):
        if temp[idx-1] < df :
            temp[idx-1] = df
    return temp
```

```python
# O(N * M)
def solution(N, A):
    temp = [0]*N
    df = 0

    for val in A:
        if val >= N+1 or (max(temp)>df)>N+1:
            df = max(temp)
        else:
            if temp[val-1] < df:
                temp[val-1] = df+1
            else:
                temp[val-1]+=1
                
    for idx in range(N):
        if temp[idx-1] < df :
            temp[idx-1] = df
            
    return temp
```
