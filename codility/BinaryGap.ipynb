```python
N = 1041
```

```python
def solution(N):
    temp = bin(N)[2:]
    temp_idx = [i for i in range(len(temp)) if temp[i]=='1']
    val = []
    
    for i in range(len(temp_idx)-1):
        val.append(abs(temp_idx[i] - temp_idx[i+1])-1)
    
    if not val:
        return 0
    return max(val)
```
